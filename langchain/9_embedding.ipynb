```python
from langchain_openai import OpenAIEmbeddings
import os
import dotenv
dotenv.load_dotenv()

embedding_model = OpenAIEmbeddings(api_key=os.getenv("openaiKey"))
```

```python
embeddings = embedding_model.embed_documents(
    ["eu gosto de cachorro",
     "eu gosto de animais"]
)
```

```python
embeddings[0]
```

```python
import numpy as np

np.dot(embeddings[0],embeddings[1])
```

```python
for i in range(len(embeddings)):
    for j in range(len(embeddings)):
        print(round(np.dot(embeddings[i],embeddings[i]),2),end=" | ")
    print()
```

```python
pergunta = "o que e um cachorro?"

embedding_query = embedding_model.embed_query(pergunta)

embedding_query
```
